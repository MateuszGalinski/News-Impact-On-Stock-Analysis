```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

## Data loading

```python
nvidia_stock = pd.read_csv("../data/data_stooq/daily/us/nasdaq stocks/2/nvda.us.txt")
nvidia_stock.head(2)
```

## Data preprocessing

```python
rename_dict = {}
for col in nvidia_stock.columns:
    rename_dict[col] = col.strip("<>")
    
# display(rename_dict)
nvidia_stock = nvidia_stock.rename(columns = rename_dict)
```

```python
display(nvidia_stock.describe())
```

```python
# DROPING OPENINT COLUMN
nvidia_stock = nvidia_stock.drop(columns=["OPENINT"])
```

```python
nvidia_stock.head(2)
```

```python
nvidia_stock["DATE"] = pd.to_datetime(nvidia_stock["DATE"], format="%Y%m%d")
print(nvidia_stock['DATE'].min())
print(nvidia_stock['DATE'].max())
```

```python
nvidia_stock.head(2)
```

```python
fig = go.Figure(data = [go.Candlestick(x=nvidia_stock['DATE'],
                        open = nvidia_stock['OPEN'],
                        high = nvidia_stock['HIGH'],
                        low = nvidia_stock['LOW'],
                        close = nvidia_stock['CLOSE'])])

fig.update_layout(
    title = dict(text = 'Ticker plot for nvidia')
)
fig.show()
```

## Calculating log return

```python
nvidia_stock["LOG_CLOSE_RETURN"] = np.log(nvidia_stock["CLOSE"] / nvidia_stock["CLOSE"].shift(1))
nvidia_stock["PCT_CLOSE_CHANGE"] = nvidia_stock["CLOSE"].pct_change() * 100
nvidia_stock.head()
```

```python
# plt.figure(figsize=(8, 4))
# plt.plot(nvidia_stock['DATE'], nvidia_stock['LOG_CLOSE_RETURN'])
# plt.show()

fig = go.Figure()
fig.add_trace(go.Scatter(x = nvidia_stock['DATE'], y =nvidia_stock['LOG_CLOSE_RETURN']))
fig.update_layout(
    title = dict(text = "Log Return plot for nvidia")
)
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = nvidia_stock['DATE'], y =nvidia_stock['PCT_CLOSE_CHANGE']))
fig.update_layout(
    title = dict(text = "Log Return plot for nvidia")
)
fig.show()
```

```python
total_log_return = nvidia_stock['LOG_CLOSE_RETURN'].sum()
total_growth_multiple = np.exp(total_log_return)
print(f"NVDA grew by a factor of {total_growth_multiple:.1f}x over the period")

print(nvidia_stock['CLOSE'].iloc[0])
print(nvidia_stock['CLOSE'].iloc[-1])
print(nvidia_stock['CLOSE'].iloc[-1]/nvidia_stock['CLOSE'].iloc[0])

# Annualized geometric daily return
n_days = nvidia_stock['LOG_CLOSE_RETURN'].count()
annualized_daily_geometric_return = np.exp(total_log_return / n_days) - 1
print(f"True (geometric) average daily return: {annualized_daily_geometric_return*100:.4f}%")
```
